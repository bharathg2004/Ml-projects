CLUSTERING ALGORITHS

1.K-MEANS CLUSTERING

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn. datasets import make_blobs
from sklearn.cluster import KMeans
from sklearn. preprocessing import StandardScaler
```

```python
X, y_true = make_blobs(n_samples=500, centers=3, cluster_std=0.60,random_state=42)
```

```python
X
```

```python
df = pd.DataFrame(X, columns=['Feature 1', 'Feature 2'])
```

```python
df
```

```python
scaler = StandardScaler()
X_scaled = scaler.fit_transform(df)
```

```python
inertia = []
k_range = range(1, 11)
```

```python
for k in k_range:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)
```

```python
inertia
```

```python
plt.plot(k_range, inertia, marker = 'o')
```

```python
kmeans_final = KMeans(n_clusters=3, random_state=42)
```

```python
cluster_labels = kmeans_final. fit_predict(X_scaled)
```

```python
cluster_labels
```

```python
df ['clusters'] = cluster_labels
```

```python
df
```

```python
sns.scatterplot(x='Feature 1', y='Feature 2', hue='clusters', data=df)
```

2.D B SCAN CLUSTERING

```python
from sklearn.datasets import make_moons
```

```python
X, y_true = make_moons(n_samples=500, noise=0.05, random_state=42)
```

```python
from sklearn.cluster import DBSCAN,KMeans
```

```python
df = pd. DataFrame(X, columns=['Feature_1', 'Feature_2' ])
```

```python
scaler = StandardScaler()
X_scaled = scaler. fit_transform(df)
```

```python
kmeans = KMeans(n_clusters=2, random_state=42)
kmeans_labels = kmeans. fit_predict(X_scaled)
```

```python
df ['kmeans_cluster' ] = kmeans_labels
```

```python
sns. scatterplot(x=df ['Feature_1'], y=df ['Feature_2' ], hue=df ['kmeans_cluster' ],
palette='tab10')
```

```python
dbscan = DBSCAN(eps=0.3, min_samples=5)
dbscan_labels = dbscan. fit_predict(X_scaled)
```

```python
df ['dbscan_cluster' ] = dbscan_labels
```

```python
sns.scatterplot(x=df['Feature_1'], y=df['Feature_2'], hue=df['dbscan_cluster'],palette='tab10')
```
